````python
top_nodes=['A','B','C','D']
bottom_nodes=[1,2,3]
B=nx.Graph()
B.add_nodes_from(top_nodes,bipartite=0)
B.add_nodes_from(bottom_nodes,bipartite=1)
edges=[('A',1),('A',2),('B',2),('C',1),('C',3),('D',3)]
B.add_edges_from(edges)
pos=nx.bipartite_layout(B,top_nodes)
nx.draw(B,pos,with_labels=True)
plt.title("Bipartite Graph")
plt.show()
````

````python
list1=["A","B","C","D"]
list2=[1,2]
G=nx.complete_bipartite_graph(list1,list2)
nx.draw_networkx(G,with_labels=True)
plt.show()
````

````python
G1=nx.Graph()
G1.add_nodes_from(['a','b','c','d','e','z'])
G1.add_edge('a','b',weight=4)
G1.add_edge('a','d',weight=2)
G1.add_edge('b','c',weight=3)
G1.add_edge('c','z',weight=2)
G1.add_edge('z','e',weight=1)
G1.add_edge('d','e',weight=3)

#length of shortest path

G1.add_edge('b','e',weight=3)
pos=nx.spring_layout(G1,iterations=50)
nx.draw_networkx(G1,pos)
label=nx.get_edge_attributes(G1,'weight')
nx.draw_networkx_edge_labels(G1,pos,edge_labels=label)
print("Shortest path:",nx.dijkstra_path(G1,'a','z'))
print("Shortest path length:",nx.dijkstra_path_length(G1,'a','z'))
````

````python
G1=nx.Graph()
G1.add_nodes_from(['a','b','c','d','e','f','g','z'])
G1.add_edge('a','c',weight=3)
G1.add_edge('a','b',weight=4)
G1.add_edge('b','d',weight=5)
G1.add_edge('d','c',weight=3)
G1.add_edge('c','e',weight=6)
G1.add_edge('d','e',weight=1)
G1.add_edge('d','f',weight=5)
G1.add_edge('e','g',weight=5)
G1.add_edge('f','g',weight=2)
G1.add_edge('f','z',weight=7)
G1.add_edge('g','z',weight=4)
#length of shortest path

G1.add_edge('a','c',weight=3)
pos=nx.spring_layout(G1,iterations=50)
nx.draw_networkx(G1,pos)
label=nx.get_edge_attributes(G1,'weight')
nx.draw_networkx_edge_labels(G1,pos,edge_labels=label)
print("Shortest path:",nx.dijkstra_path(G1,'a','z'))
print("Shortest path length:",nx.dijkstra_path_length(G1,'a','z'))
````

````python
import networkx as nx
from networkx.algorithms import tree
import matplotlib.pyplot as plt
G=nx.cycle_graph(4)
G.add_edge(0,3,weight=2)
#G.add_edge(0,1,weight=4)
mst=tree.mst.minimum_spanning_edges(G,algorithm='prim',data=False)
edgelist=list(mst)
sorted(sorted(e) for e in edgelist)
mst=nx.minimum_spanning_tree(G,algorithm='prim')
pos=nx.circular_layout(G)
nx.draw(G,pos=pos,with_labels=True)
nx.draw_networkx_edges(G,pos,edgelist=mst.edges(),width=2,edge_color='red')
plt.title("Minimum Spanning Tree")
plt.show()
````
